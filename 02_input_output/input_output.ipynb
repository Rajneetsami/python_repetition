# type coercion

```python
gravity = 9.82
mass = float(input("enter your weight in Kg: "))

force = mass*gravity

print(f"{force = }N")
```

```python
# {force = :.1f} → Special Debug Formatting
# : → start formatting
# .1f →
# f = floating-point number
# .1 = 1 digit after decimal point

print(f"{force = :.1f}N")
```

# circle geometry

```python
import math
radius = float(input("enter a radius: "))
area = math.pi * radius**2 # ** is the power operator
circumference = 2 * math.pi * radius

print(f" circle with radius {radius} has \n area: {area:.2f} a.u. \n circumference: {circumference:.2f} l.u")
```
