## 0. Pythagorean theorem

````python
# a)  A right angled triangle has the catheti: a = 3 and b = 4 length units. Compute the hypothenuse of the triangle.

import math
a = 3
b = 4

h = math.sqrt(a**2 + b**2)

print(f" hypothenuse of the the triangle is {h}")

````

````python
# b)  A right angled triangle has hypothenuse c = 7.0 and a cathetus a = 5.0 length units. Compute the other cathetus 

# and round to one decimal.

import math

a = 5.0
c = 7.0

b = math.sqrt(c**2 - a**2)

print(f"othe cathetus is {b:.1f}")
````

Calculated the accuracy using the following formula:

$\text{accuracy} = \frac{TP+TN}{TP+TN+FP+FN}$

Is this a good model? Why? Why not?

````python
TP = 2
FP = 2
FN = 11
TN = 985

accuracy = ((TP+TN)/(TP+FP+FN+TN))*100

print(f"accuracy of this model {accuracy:.2f} %")

# So even though the accuracy is 98.7%, the model is terrible at detecting fire, which is what really matters.
# Out of 13 actual fires, the model only detected 2. It missed 11 fires, which could be disastrous in a real-world scenario.
# The dataset is imbalanced
````

## 7. Theory

# a) What is the difference between input() and print() in Python?

input() is used to get input from user and print() is used to show data to the user.

# b) What data type does the input() function return by default? How can you use it to get numeric input?

By default, input() returns a string, and you can get numeric input by converting it using int() or float(), e.g., 

age = int(input("Enter age: "))

# c) Explain how formatted string literals (f-strings) improve readability in output statements. Give an example.

Formatted string literals, or f-strings, improve readability by letting you embed variables directly inside a string 

without using cumbersome concatenation (+) or multiple str() conversions.


# d) Can a variable change its data type during runtime in Python? Show an example.

Yes! In Python, a variable can change its data type at runtime because Python is dynamically typed.

x = 10

print(x, type(x))  # Output: 10 <class 'int'>

x = "Hello"

print(x, type(x))  # Output: Hello <class 'str'>

x = 3.14

print(x, type(x))  # Output: 3.14 <class 'float'>

# e) What does it mean that Python is "dynamically typed"? How is that different from statically typed languages?

Python is dynamically typed, meaning you don’t need to declare a variable’s data type—Python automatically determines 

it at runtime, unlike statically typed languages (like C or Java) where you must declare the type explicitly.
