## 5. Monte Carlo simulation (\*\*\*)

a) Simulate 5000 uniformly distributed points between -1 and 1, $X\sim U(-1,1)$. Now calculate the Euclidean distances between each point and the origin (0,0), $d_i = \sqrt{x_i^2 + y_i^2}$. For $d_i < 1$ plot the point with one color, and for $d_i > 1$ plot the point in another color, $i = \{1,\ldots, 5000\}$. (\*\*)

b) Calculate the fraction between number of inner points and number of outer points. What do you this will converge to when number of points reaches infinity? (\*\*)

````python
import random
import matplotlib.pyplot as plt
import math

N = 5000  # number of points
points_inside_x = []
points_inside_y = []
points_outside_x = []
points_outside_y = []

for _ in range(N):
    x = random.uniform(-1, 1)  # x-coordinate
    y = random.uniform(-1, 1)  # y-coordinate
    
    distance = math.sqrt(x**2 + y**2)  # distance from origin
    
    if distance < 1:
        points_inside_x.append(x)
        points_inside_y.append(y)
    else:
        points_outside_x.append(x)
        points_outside_y.append(y)

plt.figure(figsize=(6,6))
plt.scatter(points_inside_x, points_inside_y, color='green', s=5, label='Inside circle')
plt.scatter(points_outside_x, points_outside_y, color='red', s=5, label='Outside circle')
plt.xlabel("x")
plt.ylabel("y")
````

## 6. A cute rabbit among two ferocious snakes (\*\*\*)

There are three identical doors, behind two of them, there are ferocious snakes and behind one of them is a cute little rabbit. Your aim is to pick a door and get the cute rabbit. You start picking one door randomly and suddenly another door opens and snake slithers away _pssss_. Now you have a choice to either stay with your door or change to the remaining door.

<img src="https://github.com/kokchun/assets/blob/main/python/snakes_rabbits.png?raw=true" width = 200>

a) What is your choice and why? (\*\*)

b) Randomly assign which door has the rabbit, and randomly pick an initial door. Now simulate 10, 100, 1000, 10000, 100000, 1000000 times for staying and for switching doors. Calculate the proportion of finding the rabbits for each set of simulations. Plot the results (\*\*)

````python
import random
import matplotlib.pyplot as plt

# Different numbers of simulations we want to try
simulations = [10, 100, 1000, 10000, 100000, 1000000]

# Lists to store results
stay_results = []
switch_results = []

# Loop over each number of simulations
for n in simulations:
    stay_wins = 0    # count wins if we stay
    switch_wins = 0  # count wins if we switch
    
    # Repeat the game n times
    for _ in range(n):
        # Step 1: Randomly assign the rabbit behind one door
        doors = [0, 0, 0]          # 0 = snake
        rabbit_door = random.randint(0, 2)
        doors[rabbit_door] = 1     # 1 = rabbit

        # Step 2: Player picks a random door
        choice = random.randint(0, 2)

        # Step 3: Host opens a door with a snake (not the chosen door)
        remaining_doors = [i for i in range(3) if i != choice and doors[i] == 0]
        host_opens = random.choice(remaining_doors)

        # Step 4: Find the remaining unopened door
        switch_door = [i for i in range(3) if i != choice and i != host_opens][0]

        # Step 5: Count wins
        if doors[choice] == 1:       # stay
            stay_wins += 1
        if doors[switch_door] == 1:  # switch
            switch_wins += 1

    # Step 6: Store proportion of wins
    stay_results.append(stay_wins / n)
    switch_results.append(switch_wins / n)

# Step 7: Print results
print("Proportion of wins when staying:", stay_results)
print("Proportion of wins when switching:", switch_results)

# Step 8: Plot the results
plt.plot(simulations, stay_results, 'o--', label='Stay')
plt.plot(simulations, switch_results, 'o--', label='Switch')
plt.xscale('log')  # log scale for better visualization
plt.xlabel("Number of simulations")
plt.ylabel("Proportion of winning the rabbit")
plt.title("Monty Hall Simulation: Stay vs Switch")
plt.grid(True)
plt.legend()
plt.show()
````
