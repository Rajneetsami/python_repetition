## 6. Multiplication game (\*)

a) Create a multiplication game following this flow chart:

<img src="https://github.com/kokchun/assets/blob/main/python/Multiplication_game.png?raw=true">

````python
import random as rnd

score = 0
while True:
    x = rnd.randint(1, 10)
    y = rnd.randint(1, 10)
    

    find_answer = int(input(f"can you find the multiplication {x}*{y}: "))
    
    if find_answer == x*y:
        print("good work")
        score+=10
    
    else:
        print("wrong answer")
        
    ask_user = input("would you like to play again (yes/no): ")
    print(f"your final score is {score}")
    
    if ask_user != "yes":
        print("thanks player and end game")
        break

````

## Glossary

Fill in this table either by copying this into your own markdown file or copy it into a spreadsheet if you feel that is easier to work with.

| terminology           | explanation |
| --------------------- | ----------- |
| control structures    | Code that controls the flow of execution, e.g., loops and conditionals.            |
| conditional statement | A statement that executes code only if a condition is True, like if.           |
| if                    | Checks a condition and runs a block of code if True.             |
| elif                  | Stands for “else if”; checks another condition if previous if or elif were False.            |
| else                  | Runs a block of code if all previous conditions are False.            |
| for                   |A loop that runs a fixed number of times or iterates over a sequence.             |
| while                 | A loop that runs as long as a condition is True.            |
| break                 | Stops the entire loop immediately, exits the loop.            |
| continue              | Skips the current iteration and moves to the next iteration of the loop.            |
| boolean expression    | Skips the current iteration and moves to the next iteration of the loop.            |
| loop                  | A structure that repeats a block of code multiple times.            |
| nested loop           | A loop inside another loop.            |
| nested if statement   | An if statement inside another if or elif block.            |
| expression            | Produces a value, e.g., 2 + 3 or "hello".            |
| statement             | Performs an action, e.g., if, for, x = 5.            |
| control flow          | The order in which code executes, determined by loops and conditionals.           |
