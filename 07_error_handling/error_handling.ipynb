```python
# try - except

age = input("enter your age: ")

try:
    age = float(age)
    if  not 0 < age < 125:
        raise ValueError("age must be between 0 and 124")
    print(f"your age is {age} years")

except ValueError as e:
    print(f"invalid input: {e}")
```

```python
while True:
    age = input("enter your age: ")
    try:
        age = float(age)
        if  not 0 < age < 125:
            raise ValueError("age must be between 0 and 124")
        print(f"your age is {age} years")
        break

    except ValueError as e:
        print(f"invalid input: {e}")
    
```
