# Indexing

index is position number to charcters and it starts from 0. it helps to access character, string object does not support item assignment



```python
str = "apna_college"
print(str[4])
```

# slicing

accessing parts of a string

str[starting_index: ending_index]

```python
str = "Apna college"
print(str[5:len(str)])
print(str[5:]) #[5:len(str)]
print(str[:4]) # [0:4]
```

```python
# negative indexing

str = "apple"
print(str[-3:-1])
print(str[-5:-2])
```

# nesting statements

```python
age = 34

if age >= 18:
    if age >= 80:
        print("cannot drive")
    else:
        print("can drive")
else:
    print("cannot drive")
```
