# list comprehension

| Part                 | Meaning               |
| -------------------- | --------------------- |
| `x**2`               | what to store in list |
| `for x in range(10)` | loop                  |
| `if x**2 > 25`       | condition             |


```python
cubes = [x**3 for x in range(10)]
cubes
```

```python
[x**2 for x in range(10) if x**2 >25]
```

## plot graph

```python
import matplotlib.pyplot as plt

x = list(range(10))
x
```

```python
k = 2
m = 2
y = [k*x + m for x in x]
y
```

```python
plt.plot(x,y, 'o--')
plt.title("Plot y = 2x+2")
```
