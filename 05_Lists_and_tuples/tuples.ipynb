# tuples

a built in data type that lets us create immutable sequences of values.

tuple is immutable just like string

```python
tup = (87,64,33,95,76)
#tup[0]= 43
print(tup[1])
```

```python
tup = ()#empty tuple
tup = (1) # like int type
tup = (1,) # like tuple type
print(type(tup))
```

# slicing

```python
tup = (1,2,3,4,5)
print(tup[1:3])
```

# methods

```python
# tup.index(el) return of index at first occurance
# tup.count(el) count total occurence

tup = (1,2,3,4,5,2)
print(tup.index(2))
print(tup.count(2))
```

```python
# WAP to count the number of students with "A" grade in following tuple

tup = ("C", "D", "A", "A", "B", "B", "A")

print(tup.count("A"))
```

```python
# store the above values in a list & sort them from "A" to "D"

list = ["C", "D", "A", "A", "B", "B", "A"]
list.sort()
print(list)




```
