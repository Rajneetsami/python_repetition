# Lists

it can stores elements of different types(int, flost, str)

strings are immutable:  that is you can access but you can not change the value at partiular index.

lists are mutable : that is you can access as well as change the value at particular index

```python
marks = [1,2,3,4,5]
print(marks)
print(type(marks))
```

```python
print(marks[0])
print(marks[1])
print(len(marks))
```

```python
student = ["Rajneet", 34.5, 23, "Punjab"]
print(student[0])
student[0] = "Rahul"
print(student)
```

```python
string = "hello"
print(string[0])
# str[0] = 1
```

# slicing

```python
student = ["Rajneet", 34.5, 23, "Punjab"]
print(student[0:2])
print(student[:2])
print(student[1:])
print(student[-1:-3:-1])
```

# list method

```python
# list.append(4) add one element at the end

list = [2,1,3,5]
print(list.append(4)) # append() modifies the original list in memory so doesn’t create a new list → returns None.
#list
```

```python
# list.sort() sort in asecendind order
# list.sort(reverse = True) sort in descending order

list.sort()
list.sort(reverse = True)
list
```

```python
# list.insert(idx, el) insert element at index

num = [1,2,3,4,5]

num.insert(3, 5)
num
```

```python
# list. remove(1) remove firts accourence of the element
# list.pop(idx) remove element at particular index

num = [1,2,3,4,5]
num.remove(1)
num.pop(3)
num

```
