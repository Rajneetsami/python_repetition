# sets

collection of unordered items, set are mutable

each element in the set must be unique and immutable

acceptable data types int, float, tuple, str, bool not dict and list



```python
collection = {1,2,2,2, "hello", "world", "world",4}
print(len(collection))
```

```python
num = set() # empty set
```

# methods

```python
# set.add(el)
# set.remove(el) 
# set.clear() # empty set
# set.pop() # removes random value

num.add(1)
num.add(2)
num.add(3)
num.add(3)
num.add("Rajneet kaur")
num.add((1,2,3))

num.remove(1)
num

num.pop()
num.clear()
num
```

```python
# set.union(set2) combines both set values & returns new
# set.intersection(set2) combine common values & return new

set1 = {1,2,3}
set2 = {3,4,5}
print(set1.union(set2))
print(set1.intersection(set2))
```
