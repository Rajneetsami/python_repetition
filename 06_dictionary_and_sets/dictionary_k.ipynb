# dictionary comprehension

```python
# create grade with different limits

grade_limlit = {grade: i*10 for i, grade in enumerate("FEDCBA")}
grade_limlit
```

```python
import random as rnd

def generate_scores(number_scores):
    rnd.seed(32)
    scores = [rnd.randint(0,60) for _ in range(number_scores)]
    return scores

my_scores = generate_scores(20)
print(my_scores)
```

```python
grade_count = {key: 0 for key in grade_limlit}
grade_count
```

```python
for score in my_scores:
    for grade, limit in grade_limlit.items():
        if limit <= score < limit + 10:
            grade_count[grade]  += 1

print(f"grade count {grade_count}")
            
    
```

1️⃣ Define Grade Boundaries

Each grade (A, B, C, D, E, F) corresponds to a score range.
For example:

F → 0–9
E → 10–19
D → 20–29
…
A → 50–59
Logic: This tells us which score maps to which grade.

2️⃣ Generate Scores
You have a list of students’ scores.
The scores can be random or real data.
Logic: We need a set of numbers to see how students perform.

3️⃣ Initialize Counters
For each grade, you need a counter starting at 0.
Logic: This is to track how many students fall into each grade.


4️⃣ Assign Scores to Grades
For each student’s score:
Compare the score with the grade ranges.
Determine which grade the score belongs to.
Increment the counter for that grade.
Logic: This step classifies students into their respective grades and keeps count.

5️⃣ Summarize Results
After all scores are processed, you have the number of students per grade.
Logic: This gives you a distribution of grades across the class.
