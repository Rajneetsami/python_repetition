# mathematic set operators

-union

-intersection

-except

```python
first_month_customer = {"alice", "bob", "carla", "daniel"}
second_month_customer = {"carla", "daniel", "elena"}

returning_customer =first_month_customer.intersection(second_month_customer)
returning_customer
```

```python
all_customers =first_month_customer.union(second_month_customer)
all_customers
```

```python
one_time_customer = first_month_customer.union(second_month_customer) - first_month_customer.intersection(second_month_customer)
one_time_customer
```
