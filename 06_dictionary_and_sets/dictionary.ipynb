# nested dictionary

```python
student = {"name": "Rajneet Kaur",
"subjects": { "phy": 97,
"maths": 98,
"chem": 95

}
}

student["subjects"]["maths"]
```

# methods

```python
# mydict.keys()
# mydict.values()

print(list(student.keys()))
print(len(student))
print(list(student.values()))
```

```python
# my.dict.items return all pairs as tuples

pairs = list(student.items())

pairs[0]
```

```python
# mydict.get("key") return value according to value

print(student.get("name1")) # this will return none
print(student["name1"])# this will return error
```

```python
# mydict.update(newdict)

student = {"name": "Rajneet Kaur",
"subjects": { "phy": 97,
"maths": 98,
"chem": 95

}
}
student.update({"city":"punjab", "age": 16})

student
```
