# while loops

used to repeat instruction


```python
str = "hello"
count = 1 # iterator

while count <= 100: # iteration
    print(str, count)
    count+=1
print(count)
```

```python
i = 5

while i >= 1:
    print(i)
    i-=1
print("ended")
```

```python
# print number from 1 to 100

num = 1

while num <= 100:
    print(f" number is {num}")
    num+=1
    
```

```python
# print numbers from 100 to 1

num = 100

while num >=1:
    print(f"number is {num}")
    num-=1
    
    
```

```python
# print multiplication table of a number n

n = int(input("enter number: "))
i = 1

while i <= 10:
    print(n*i)
    i+=1
```

```python
# print the element of the following list using a loop
# [1,4,9,16,25,36,49,64,81,100]

i = 1

while i <= 10:
    print(i**2)
    i+=1
    
```

```python
# print the element of the following list using a loop
nums =  [1,4,9,16,25,36,49,64,81,100]
idx = 0

# traverse
while idx < len(nums):
    print(nums[idx])
    idx+=1
    
```

```python
# search for a number x in this tuple using loop

tup = (1,4,9,16,25,36,49,64,81,100)
idx = 0 # intilaization

x = int(input("enter number: "))

while idx < len(tup):

    if tup[idx] == x:
        print("found")
        break
    else:
        print("finding")
    idx+=1
    
print("end of loop")  

```

## break and continue


```python
i = 1

while i <= 5:
    print(i)
    if (i == 3):
        break
    i+=1
print("end of loop")   
```

```python
i = 0

while i <= 10:
    if (i % 2 == 0):
        i+=1
        continue # act as skip
    print(i)
    i+=1
print("end of loop")
```

```python
# WAP to find the sum of first n natural numbers

num = int(input("enter number : "))

sum = 0
i = 1

while i <= num:
    sum+=i
    i += 1
    
print("total sum: ", sum)
    
```
