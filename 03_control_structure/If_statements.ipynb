## logical operators

Name	     Symbol	    Example	   Result

Less than	   <	    3 < 2	    False

Less than or equal to	<=	3 <= 3	True

Larger than	   >	    3 > 2	    True

Larger than or equal to	>=	-3 >= 4	False

Equal to	   ==	    3 == 3	    True

Not equal	   !=	    3 != 2	    True

And operator   and	  3 == 2 and 3 > 2	False

Or operator	   or	   3 == 2 or 3 > 2	True

Not operator	not	   not 3 == 2	 True

# truthy and falsy values

In Python, some values are automatically considered False.

These are called falsy values.

âŒ Falsy values:

0

0.0

"" (empty string)

[] (empty list)

{} (empty dictionary)

() (empty tuple)

None

False

```python
number = 5

if not number % 2:
    print("even number")
else:
    print ("odd number")
```

# nested if statements 

```python
number = 25
if number % 5 == 0:
    print (f"{number} is divisible by 5")
    if number % 2 == 0 :
        print(f"{number} is even")
    else:
        print(f"{number} is odd")
```
