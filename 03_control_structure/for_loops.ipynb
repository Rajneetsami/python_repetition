## for loops

for traversing lsit, string, tuples

```python
list = [1,2,3,4,5]

for element in list:
    print(element)
```

```python
veggies = ["potato", "brinjal", "ladyfinger", "cucumber"]

for value in veggies:
    print(value)
```

```python
tup = (2,4,6,8,10)

for num in tup:
    print(num)
```

```python
string = "apna college"

for cha in string:
    if (cha == "o"):
        print("Found")
        break
    print(cha)
    
else:
    print("end")
```

```python
# print element of the following list using a loop
list = [1,4,9,16,25,36,49,64,81,100]

for element in list:
    print(element)

```

```python
# search for a number x in this tuple using loop:
tup = (1,4,9,16,25,36,49,64,81,64)

x = int(input("enter number: "))

idx = 0

for el in tup: # linear search
    if el == x:
        print("matched at", idx)
        break
    idx+=1
```

## range()

```python
print(range(5))
```

```python
for i in range(5): # stop condition
    print(i)
```

```python
for i in range(2,10): # start and stop
    print(i)
```

```python
for i in range(2,10,2): # start, stop and step 
    print(i)
```

```python
for i in range(2,100,2):
    print(i)
```

```python
# print number from 1 to 100:

for i in range(1,101):
    print(i)
```

```python
# print number from 100 to 1

for i in range(100,0,-1):
    print(i)
```

```python
# print the multiplication table of number n

n = int(input("enter number: "))

for i in range(11):
    print(n*i)
```

## pass statement


```python
for i in range(5):
    pass

if i > 5:
    pass

print("ended loop")

```

```python
# WAP to find the factorial of first n numbers

n = int(input("enter number : "))
fac = 1

for i in range(1, n+1):
    fac*=i
print(fac)
    
```
